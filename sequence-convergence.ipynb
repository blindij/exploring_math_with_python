```python
import numpy as np
import matplotlib.pyplot as plt

# Consider up to n = 40
n = np.arange(1, 41)
recp = 1/n
# Define x_n
xn = 1 - recp
# Error = |x_n - x| where x = lim x_n = 0  when n -> infinity
err_xn = recp
# lim y_n = 0 when n -> infinity
yn = np.sin(n)/n
err_yn = abs(yn)

# lim zn = 0 when n -> infinity
zn = np.log(n)/n
err_zn = abs(zn)

# lim E_n = e when n -> infinity
En = (1 + recp)**n
err_En = abs(En - np.e)

# Plotting with dots (markersize = 3) joined by various line styles
plt.plot(n, err_xn, 'o-', ms=3)
plt.plot(n, err_yn, 'o--', ms=3)
plt.plot(n, err_zn, 'o-.', ms=3)
plt.plot(n, err_En, 'o:', ms=3)

plt.xlim(0, 40)
plt.ylim(0, 0.4)
plt.xlabel(r'$n$')
plt.ylabel('|Error|')
plt.legend([r'$|x_n-1|$', r'$|y_n|$', r'$|z_n|$', r'$|E_n-e|$'])
plt.grid('on')
plt.show()
```
