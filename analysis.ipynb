### Finding N given e

```python
# For any given e, say, 10-5
epsilon = 1e-5
n = 1
err = np.e - (1 + 1/n)**n
# Continue searching until the error < e
# Increasing n by 1 per iteration
while err >= epsilon:
    n += 1
    err = np.e - (1 + 1/n)**n

print("N_min = ", n - 1)
```
